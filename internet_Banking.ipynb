```python
print("\033[1m-UNIVILLE Internet Banking-\033[0m")  # esse \033[1m - \033[0m  é para deixar o texto do print em negrito, achei no google
print("1-Acessar")
print("2-Encerrar \n")

escolha = int(input("Esolha a opção: "))

if escolha == 1:


  tentativas = 3
  # while para acesso ao sistema.
  while tentativas > 0:
    usuario = input("\nDigite o seu usuário: ")
    senha = input("Digite a senha: ")
    #verifica se o acesso foi permitido
    if usuario != 'leonardo' or senha != '123':
      print("Usuário ou senha inválidos!")
      tentativas -= 1
      print("Tentativas restantes:",tentativas)
    else:
      print("\033[1m\nBem vindo",usuario,"\033[0m")
      print("\n\033[1m-UNIVILLE Internet Banking-\033[0m")
      print("1. Consultar Saldo")
      print("2. Realizar Saque")
      print("3. Realizar Depósito")
      print("4. Consultar limite")
      print("5. Encerrar")

      saldo = 0

      # segunda escolha, agora já dentro do sistema
      while True: # Loop infinito até o usuário escolher encerrar
        escolha2 = int(input("\nEscolha o que fazer: "))

        if escolha2 == 5:
          print("\033[1m\nAté mais\033[0m")
          break # Sai do loop interno e, consequentemente, do sistema
        elif escolha2 == 3:
          deposito = float(input("Digite o valor do depósito: "))
          saldo += deposito
          print(f"Depósito de R$ {deposito:.2f} realizado com sucesso. Novo saldo: R$ {saldo:.2f}")
        elif escolha2 == 1:
          print(f"Saldo R$:{saldo:.2f}")
        elif escolha2 == 2:
          saque = float(input("Digite o valor do saque: "))
          if saque > saldo:
            print("Saque inválido! Valor maior que o saldo")
          else:
            saldo -= saque
            print(f"Saque de R$ {saque:.2f} realizado com sucesso. Novo saldo: R$ {saldo:.2f}")
        else:
          print("Opção inválida. Por favor, escolha novamente.")

      break # Sai do loop de tentativas após login bem-sucedido e operações


  # negação do acesso
  else:
    print("\033[1mTentativas acabaram, acesso bloqueado!\033[0m")
else:
  print("Até mais")
  exit()
```
